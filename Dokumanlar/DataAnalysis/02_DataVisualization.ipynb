## Matplotlib

Basit Grafik Olu≈üturma

```python
import matplotlib.pyplot as plt
import numpy as np

# Veri olu≈üturma
aylar = ['Ocak', '≈ûubat', 'Mart', 'Nisan', 'Mayƒ±s', 'Haziran']
satis_miktarlari = [120, 150, 130, 180, 200, 220]

# Grafik olu≈üturma
plt.figure(figsize=(10, 6))  # Grafik boyutunu ayarlama
plt.plot(aylar, satis_miktarlari, marker='o', linestyle='-', color='b', linewidth=2)

# Grafik ba≈ülƒ±klarƒ± ve etiketler
plt.title('Aylƒ±k Satƒ±≈ü Miktarlarƒ±', fontsize=16)
plt.xlabel('Aylar', fontsize=12)
plt.ylabel('Satƒ±≈ü Miktarƒ± (Adet)', fontsize=12)

# Izgara eklemek okunabilirliƒüi artƒ±rƒ±r
plt.grid(True, linestyle='--', alpha=0.7)

# Grafik g√∂sterimi
plt.show()
```

Kategori Kar≈üƒ±la≈ütƒ±rma

```python
import matplotlib.pyplot as plt

# Veri olu≈üturma
urunler = ['√úr√ºn A', '√úr√ºn B', '√úr√ºn C', '√úr√ºn D']
satis_adetleri = [240, 180, 300, 210]
ortalama_fiyat = [150, 200, 120, 180]

# Grafik olu≈üturma
plt.figure(figsize=(12, 7))

# ƒ∞lk y ekseni - Satƒ±≈ü adetleri
ax1 = plt.subplot()
ax1.bar(urunler, satis_adetleri, color='skyblue', alpha=0.7, label='Satƒ±≈ü Adedi')
ax1.set_xlabel('√úr√ºnler', fontsize=12)
ax1.set_ylabel('Satƒ±≈ü Adedi', fontsize=12, color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

# ƒ∞kinci y ekseni - Ortalama fiyat
ax2 = ax1.twinx()
ax2.plot(urunler, ortalama_fiyat, color='darkred', marker='s', linewidth=2, label='Ortalama Fiyat')
ax2.set_ylabel('Ortalama Fiyat (TL)', fontsize=12, color='darkred')
ax2.tick_params(axis='y', labelcolor='darkred')

# Ba≈ülƒ±k ve legend
plt.title('√úr√ºn Bazƒ±nda Satƒ±≈ü Adedi ve Ortalama Fiyat Kar≈üƒ±la≈ütƒ±rmasƒ±', fontsize=14)
fig = plt.gcf()  # Mevcut fig√ºr√º al
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)

# Grafik g√∂sterimi
plt.tight_layout()
plt.show()
```

## Seaborn

Histogram ve Yoƒüunluk Grafiƒüi

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Rastgele veri olu≈üturma (m√º≈üteri ya≈ülarƒ±)
np.random.seed(42)
musteri_yaslari = np.random.normal(loc=35, scale=10, size=1000)
musteri_yaslari = np.clip(musteri_yaslari, 18, 70)  # Ya≈ülarƒ± 18-70 aralƒ±ƒüƒ±nda sƒ±nƒ±rla

# DataFrame olu≈üturma
df = pd.DataFrame({'Ya≈ü': musteri_yaslari})

# Grafik olu≈üturma
plt.figure(figsize=(12, 7))

# Histogram ve yoƒüunluk grafiƒüi
sns.histplot(data=df, x='Ya≈ü', bins=30, kde=True, color='teal', alpha=0.7)

# Grafik √∂zelliklerini ayarlama
plt.title('M√º≈üteri Ya≈ü Daƒüƒ±lƒ±mƒ±', fontsize=16)
plt.xlabel('Ya≈ü', fontsize=12)
plt.ylabel('Frekans', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ortalama ve medyan √ßizgileri ekleme
ortalama = df['Ya≈ü'].mean()
medyan = df['Ya≈ü'].median()
plt.axvline(ortalama, color='red', linestyle='dashed', linewidth=2, label=f'Ortalama: {ortalama:.1f}')
plt.axvline(medyan, color='green', linestyle='dashed', linewidth=2, label=f'Medyan: {medyan:.1f}')

plt.legend()
plt.show()
```

Kutu Grafiƒüi (Box Plot) ve Violin Grafiƒüi

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Farklƒ± departmanlarda √ßalƒ±≈üanlarƒ±n maa≈ü verisi
np.random.seed(42)
departmanlar = ['Yazƒ±lƒ±m', 'Pazarlama', 'Satƒ±≈ü', 'ƒ∞nsan Kaynaklarƒ±', 'Muhasebe']
maaslar = []

for dept in departmanlar:
    if dept == 'Yazƒ±lƒ±m':
        maaslar.extend(np.random.normal(loc=15000, scale=3000, size=50))
    elif dept == 'Pazarlama':
        maaslar.extend(np.random.normal(loc=10000, scale=2000, size=50))
    elif dept == 'Satƒ±≈ü':
        maaslar.extend(np.random.normal(loc=12000, scale=2500, size=50))
    elif dept == 'ƒ∞nsan Kaynaklarƒ±':
        maaslar.extend(np.random.normal(loc=9000, scale=1500, size=50))
    else:  # Muhasebe
        maaslar.extend(np.random.normal(loc=11000, scale=2000, size=50))

df = pd.DataFrame({
    'Departman': np.repeat(departmanlar, 50),
    'Maa≈ü': maaslar
})

# Grafik olu≈üturma
plt.figure(figsize=(14, 8))

# Kutu grafiƒüi
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='Departman', y='Maa≈ü', palette='Set2')
plt.title('Departmanlara G√∂re Maa≈ü Daƒüƒ±lƒ±mƒ± (Kutu Grafiƒüi)', fontsize=14)
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Violin grafiƒüi
plt.subplot(1, 2, 2)
sns.violinplot(data=df, x='Departman', y='Maa≈ü', palette='Set2', inner='quartile')
plt.title('Departmanlara G√∂re Maa≈ü Daƒüƒ±lƒ±mƒ± (Violin Grafiƒüi)', fontsize=14)
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

Korelasyon Isƒ± Haritasƒ±

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# E-ticaret veri seti olu≈üturma
np.random.seed(6)
musteri_sayisi = 500

veri = {
    'G√ºnl√ºk_Ziyaret_Sayisi': np.random.poisson(lam=5, size=musteri_sayisi),
    'Ortalama_Sayfa_Suresi': np.random.gamma(shape=2, scale=30, size=musteri_sayisi),
    'Sepete_Eklenen_Urun_Sayisi': np.random.poisson(lam=2, size=musteri_sayisi),
    'Satƒ±n_Alma_Sayisi': np.random.poisson(lam=0.5, size=musteri_sayisi),
    'Harcama_Miktari': []
}

# Harcama miktarƒ±nƒ± satƒ±n alma sayƒ±sƒ±na baƒülƒ± olarak hesapla
for i in range(musteri_sayisi):
    if veri['Satƒ±n_Alma_Sayisi'][i] > 0:
        veri['Harcama_Miktari'].append(np.random.normal(loc=250, scale=50) * veri['Satƒ±n_Alma_Sayisi'][i])
    else:
        veri['Harcama_Miktari'].append(0)

df = pd.DataFrame(veri)

# Korelasyon matrisini hesapla
korelasyon_matrisi = df.corr()

# Grafik olu≈üturma
plt.figure(figsize=(12, 10))

# Isƒ± haritasƒ±
sns.heatmap(korelasyon_matrisi,
            annot=True,
            fmt=".2f",
            cmap='coolwarm',
            linewidths=0.5,
            annot_kws={"size": 12})

plt.title('E-Ticaret Metrikleri Arasƒ±ndaki Korelasyon', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Zaman serisi verisi olu≈üturma
tarihler = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
np.random.seed(42)
satislar = np.cumsum(np.random.normal(loc=10, scale=5, size=len(tarihler))) + 100

# DataFrame olu≈üturma
df = pd.DataFrame({
    'Tarih': tarihler,
    'G√ºnl√ºk Satƒ±≈ü': satislar
})

# Etkile≈üimli √ßizgi grafiƒüi
fig = px.line(df,
              x='Tarih',
              y='G√ºnl√ºk Satƒ±≈ü',
              title='2023 Yƒ±lƒ± G√ºnl√ºk Satƒ±≈ü Performansƒ±',
              labels={'G√ºnl√ºk Satƒ±≈ü': 'Satƒ±≈ü Miktarƒ±', 'Tarih': 'Tarih'},
              template='plotly_dark')

# Grafik √∂zelliklerini √∂zelle≈ütirme
fig.update_traces(line=dict(width=2.5, color='#1f77b4'))
fig.update_layout(
    hovermode='x unified',
    xaxis_title='Tarih',
    yaxis_title='Satƒ±≈ü Miktarƒ±',
    font=dict(family="Arial", size=12),
    title_font_size=18,
    hoverlabel=dict(bgcolor="white", font_size=14)
)

# Trend √ßizgisi ekleme
fig.add_scatter(
    x=df['Tarih'],
    y=df['G√ºnl√ºk Satƒ±≈ü'].rolling(window=30).mean(),
    mode='lines',
    name='30 G√ºnl√ºk Ortalama',
    line=dict(color='red', width=2, dash='dash')
)

fig.show()
```

Etkile≈üimli Daƒüƒ±lƒ±m Grafiƒüi (Scatter Plot)

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Yapay veri olu≈üturma - E-ticaret kullanƒ±cƒ±larƒ±
np.random.seed(42)
kullanici_sayisi = 300

veri = {
    'Ziyaret_Sayisi': np.random.poisson(lam=8, size=kullanici_sayisi),
    'Ortalama_Sure': np.random.gamma(shape=3, scale=25, size=kullanici_sayisi),
    'Sepet_Ekleme': np.random.poisson(lam=3, size=kullanici_sayisi),
    'Satƒ±n_Alma': np.random.binomial(n=1, p=0.3, size=kullanici_sayisi),
    'Cinsiyet': np.random.choice(['Erkek', 'Kadƒ±n'], size=kullanici_sayisi, p=[0.48, 0.52])
}

df = pd.DataFrame(veri)

# Satƒ±n alma olasƒ±lƒ±ƒüƒ±nƒ± hesapla
df['Satƒ±n_Alma_Olasiligi'] = df['Satƒ±n_Alma'].map({0: 'Satƒ±n Alma Yok', 1: 'Satƒ±n Alma Var'})

# Etkile≈üimli daƒüƒ±lƒ±m grafiƒüi
fig = px.scatter(df,
                 x='Ziyaret_Sayisi',
                 y='Ortalama_Sure',
                 size='Sepet_Ekleme',
                 color='Satƒ±n_Alma_Olasiligi',
                 hover_name='Cinsiyet',
                 title='Kullanƒ±cƒ± Davranƒ±≈ü Analizi',
                 labels={
                     'Ziyaret_Sayisi': 'Aylƒ±k Ziyaret Sayƒ±sƒ±',
                     'Ortalama_Sure': 'Ortalama Sayfa S√ºresi (saniye)',
                     'Sepet_Ekleme': 'Sepete Eklenen √úr√ºn Sayƒ±sƒ±'
                 },
                 template='plotly_white')

# Grafik d√ºzenini iyile≈ütirme
fig.update_traces(marker=dict(line=dict(width=1, color='DarkSlateGrey')),
                  selector=dict(mode='markers'))
fig.update_layout(
    legend_title_text='Satƒ±n Alma Durumu',
    font=dict(family="Arial", size=12),
    title_font_size=18
)

fig.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

# T√ºrkiye illerine g√∂re yapay satƒ±≈ü verisi
iller = ['ƒ∞stanbul', 'Ankara', 'ƒ∞zmir', 'Bursa', 'Antalya', 'Adana', 'Konya', 'Gaziantep',
         'Mersin', 'Samsun', 'Diyarbakƒ±r', 'Kayseri', 'Eski≈üehir', 'Trabzon', 'Malatya']
np.random.seed(42)
satis_miktarlari = np.random.randint(50, 500, size=len(iller))

# DataFrame olu≈üturma
df = pd.DataFrame({
    'ƒ∞l': iller,
    'Satƒ±≈ü_Miktarƒ±': satis_miktarlari
})

# T√ºrkiye harita verisi (basitle≈ütirilmi≈ü koordinatlar)
turkiye_harita = {
    'ƒ∞l': ['ƒ∞stanbul', 'Ankara', 'ƒ∞zmir', 'Bursa', 'Antalya', 'Adana', 'Konya', 'Gaziantep',
           'Mersin', 'Samsun', 'Diyarbakƒ±r', 'Kayseri', 'Eski≈üehir', 'Trabzon', 'Malatya'],
    'Enlem': [41.0082, 39.9334, 38.4192, 40.1825, 36.8969, 36.9916, 37.8700, 37.0662,
              36.8121, 41.2992, 37.9132, 38.7326, 39.7667, 41.0027, 38.3580],
    'Boylam': [28.9784, 32.8597, 27.1287, 29.0669, 30.7133, 35.3250, 32.4824, 37.3833,
               34.6415, 36.3356, 40.2170, 35.4852, 30.5256, 39.7163, 38.0360]
}

harita_df = pd.DataFrame(turkiye_harita)
df = pd.merge(df, harita_df, on='ƒ∞l')

# Etkile≈üimli harita
fig = px.scatter_mapbox(df,
                        lat='Enlem',
                        lon='Boylam',
                        size='Satƒ±≈ü_Miktarƒ±',
                        color='Satƒ±≈ü_Miktarƒ±',
                        hover_name='ƒ∞l',
                        hover_data=['Satƒ±≈ü_Miktarƒ±'],
                        color_continuous_scale=px.colors.sequential.Viridis,
                        size_max=50,
                        zoom=5,
                        title='T√ºrkiye ƒ∞llerine G√∂re Satƒ±≈ü Daƒüƒ±lƒ±mƒ±',
                        mapbox_style='carto-positron')

fig.update_layout(
    margin={"r":0,"t":30,"l":0,"b":0},
    coloraxis_colorbar=dict(title="Satƒ±≈ü Miktarƒ±")
)

fig.show()
```
