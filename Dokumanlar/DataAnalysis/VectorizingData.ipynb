```python
import numpy as np

# Geleneksel döngü ile getiri hesaplama
prices = [100, 102, 101, 105, 107]  # Örnek fiyatlar
returns = []
for i in range(1, len(prices)):
    returns.append((prices[i] - prices[i-1]) / prices[i-1])
returns
```

```python
# RGB görüntüsünü temsil eden 3D dizi (yükseklik, genişlik, kanal)
image_rgb = np.random.randint(0, 256, (100, 100, 3), dtype=np.uint8)

# Gri seviye dönüşüm formülü: Y = 0.2989*R + 0.5870*G + 0.1140*B
image_gray = (
    0.2989 * image_rgb[:, :, 0] +
    0.5870 * image_rgb[:, :, 1] +
    0.1140 * image_rgb[:, :, 2]
).astype(np.uint8)
```

```python
import matplotlib.pyplot as plt
plt.imshow(image_gray)
image_gray.shape,image_rgb.shape
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(np.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

vectorized_prime = np.vectorize(is_prime)
numbers = np.array([2, 3, 4, 5, 6, 7])
results = vectorized_prime(numbers)  # [True, True, False, True, False, True]
results
```
