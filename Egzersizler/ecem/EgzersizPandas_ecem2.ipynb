```python
## Boş verilerin temizlenmesi
medyanage = data["Age"].median()
data["Age"] = data["Age"].fillna(medyanage)
data.drop(["Cabin"],axis=1,inplace=True)
modeEmbarked = data["Embarked"].mode()[0]
data["Embarked"] = data["Embarked"].fillna(modeEmbarked)
```
