```python
yol2 = "../../Egzersizler/datasets/titanic/train.csv"
```

```python
titanic_df = pd.read_csv(yol2)
```

```python
titanic_df.head()
```

```python
titanic_df.isna()
```

```python
titanic_df.isnull().sum()
```

```python
median_age = titanic_df['Age'].median()
median_age
```

```python
mode_embarked = titanic_df['Embarked'].mode()[0]
```

```python
titanic_df['Age'] = titanic_df['Age'].fillna(median_age)
titanic_df.drop(['Cabin'], axis=1, inplace=True)
titanic_df['Embarked'] = titanic_df['Embarked'].fillna(mode_embarked)
```

```python
titanic_df.isnull().sum()
```

Cinsiyete göre hayatta kalma oranı

```python
import matplotlib.pyplot as plt
titanic_df['Survived']
```

```python
# Grafik oluşturma
plt.bar(titanic_df['Sex'], titanic_df['Survived'].sum())

# Grafik başlıkları ve etiketler
plt.title('Cinsiyete göre hayatta kalma oranı', fontsize=16)
plt.xlabel('Cinsiyet', fontsize=12)
plt.ylabel('Hayatta kalma oranı', fontsize=12)

# Grafik gösterimi
plt.show()
```
